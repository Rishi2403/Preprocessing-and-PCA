```python
X_scaledd=scaled_df.drop(['Classes'],axis=1)
```

```python
X_scaledd
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X_scaledd,y,test_size=0.33,random_state=42)
```

```python
# Implement PCA on the preprocessed dataset using the scikit-learn library.

pca = PCA(n_components=8)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
# prompt: Determine the optimal number of principal components to retain based on the explained variance ratio.

pca = PCA()
pca.fit(X_scaledd)
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio vs. Principal Component')
plt.show()
```

```python
#  Visualise the results of PCA using a scatter plot.

import matplotlib.pyplot as plt

plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train.values)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Scatter plot of PCA results')
plt.show()
```
